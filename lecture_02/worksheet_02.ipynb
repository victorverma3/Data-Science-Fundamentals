a) What is a drawback of the top down approach?

The top down approach can lead to a longer debugging process, both in terms of time and effort.

b) What is a drawback of the bottom up approach?

It requires a lot more planning, and there could be some time wasted if some functions end up not being useful, or if it wasn't planned properly.

c) What are 3 things you can do to have a better debugging experience?

1. Don't panic.
2. Read the error (What is it telling you? Where did it occur? Is it a cause or a symptom?).
3. Re-read your code (Can you mentally trace through the code to reproduce the error?).
4. Sanity check the workspace.
5. Look online for help.

```python
class Board:
    def __init__(self):
        self.board = [["-" for _ in range(8)] for _ in range(8)]
    def __repr__(self):
        res = ""
        for row in range(8):
            for col in range(8):
                res += self.board[row][col]
                res += " "
            res += "\n"
        return res
    def set_queen_at(self, row, col):
        self.board[row][col] = "Q"
    def unset_queen_on_row(self, row):
        self.board[row] = ["-" for _ in range(8)]
    def find_solution(self):
        row, col = 0, 0
        while row < 8:
            # we are searching for a solution
            if self.is_valid_move(row, col):
                self.set_queen_at(row, col)
                row += 1
                col = 0
            else:
                col += 1
                if col >= 8:
                    # we weren't able to place a queen on this row
                    # we need to backtrack and adjust the position
                    # of the queen on the previous row
                    col = self.get_queen_on_row(row - 1)
                    col += 1
                    row -= 1
        # we have found a solution
        print("Found a solution: ")
        print(self)

b = Board()
b.set_queen_on(1, 1)
print(b)
b.unset_queen_on_row(1)
print(b)
```

d) Complete the code below to replaces `male` with 0 and `female` with 1

```python
df['Sex'] = df["Sex"].replace({"male": 0, "female": 1})
df.head()
```

e) Complete the code below to add four columns `First Name`, `Middle Name`, `Last Name`, and `Title` corresponding to the value in the `name` column.

For example: `Braund, Mr. Owen Harris` would be: 

|First Name | Middle Name | Last Name | Title |

| Owen      |  Harris     |  Braund   | Mr    |

Anything not clearly one of the above 4 categories can be ignored.

```python

df[['First Name', 'Middle Name', 'Last Name', 'Title']] = df['Name'].str.split(' ', expand=True)[[2, 3, 0, 1]].apply(lambda x: x.str.rstrip(','))
df.head()
```
